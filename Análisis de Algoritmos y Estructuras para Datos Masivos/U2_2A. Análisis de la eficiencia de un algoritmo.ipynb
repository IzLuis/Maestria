Utilizar el notebook de Jupyter para generar las siguientes comparaciones de ordenes de crecimiento (una figura por comparación, i.e., cinco figuras). Escoja los rangos adecuados para cada comparación, ya que como será evidente después, no es práctico fijar los rangos.

- 𝑂(1) 𝑣𝑠 𝑂(log𝑛)

- 𝑂(𝑛) 𝑣𝑠 𝑂(𝑛log𝑛)

- 𝑂(𝑛2) 𝑣𝑠 𝑂(𝑛3)

- 𝑂(𝑎𝑛) 𝑣𝑠 𝑂(𝑛!)

- 𝑂(𝑛!) 𝑣𝑠 𝑂(𝑛𝑛)

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd
```

## - O(1) 𝑣𝑠 O(log(n))

```python
x = np.arange(1, 1000)
y1 = np.ones_like(x)
y2 = np.log(x)

plt.plot(x, y1, '-', label='O(1)')
plt.plot(x, y2, '-', label='O(log(n))')
plt.xlabel('x')
plt.ylabel('y')
plt.title('O(1) vs O(logn)')
plt.legend()
plt.grid(True)
plt.show()
```

## - 𝑂(𝑛) 𝑣𝑠 𝑂(𝑛log(𝑛))

```python
x = np.arange(1, 1000)
y1 = x
y2 = x * np.log(x)

plt.plot(x, y1, '-', label='O(n)', color='red')
plt.plot(x, y2, '-', label='O(nlog(n))', color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title('O(n) vs O(nlogn)')
plt.legend()
plt.grid(True)
plt.show()
```

## - 𝑂(𝑛^2) 𝑣𝑠 𝑂(𝑛^3)

```python
x = np.arange(1, 10)
y1 = np.power(x, 2)
y2 = np.power(x, 3)

plt.plot(x, y1, '-', label='O(n^2)', color='purple')
plt.plot(x, y2, '-', label='O(n^3)', color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title('O(n^2) vs O(n^3)')
plt.legend()
plt.grid(True)
plt.show()
```

## - 𝑂(𝑎^𝑛) 𝑣𝑠 𝑂(𝑛!)

```python
x = np.arange(1, 11)
a = 4
y1 = np.power(a, x)
y2 = [math.factorial(n) for n in x]

plt.plot(x, y1, '-', label='O(a^n)', color = 'orange')
plt.plot(x, y2, '-', label='O(n!)', color = 'blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('O(a^n) vs O(n!)')
plt.legend()
plt.grid(True)
plt.show()
```

## - 𝑂(𝑛!) 𝑣𝑠 𝑂(𝑛^𝑛)

```python
x = np.arange(1, 6)
y1 = [math.factorial(n) for n in x]
y2 =  np.power(x, x)

plt.plot(x, y1, '-', label='O(n!)', color = "blue")
plt.plot(x, y2, '-', label='O(n^n)', color = "red")
plt.xlabel('x')
plt.ylabel('y')
plt.title('O(n!) vs O(n^n)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def factorial (n):
    if n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n+1):
            fact *= i
        return fact
```
