```python
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd
```

## - O(1) ùë£ùë† O(log(n))

```python
x = np.arange(1, 1000)
y1 = np.ones_like(x)
y2 = np.log(x)

plt.plot(x, y1, '-', label='O(1)')
plt.plot(x, y2, '-', label='O(log(n))')
plt.xlabel('x')
plt.ylabel('y')
plt.title('O(1) vs O(logn)')
plt.legend()
plt.grid(True)
plt.show()
```

## - ùëÇ(ùëõ) ùë£ùë† ùëÇ(ùëõlog(ùëõ))

```python
x = np.arange(1, 1000)
y1 = x
y2 = x * np.log(x)

plt.plot(x, y1, '-', label='O(n)', color='red')
plt.plot(x, y2, '-', label='O(nlog(n))', color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title('O(n) vs O(nlogn)')
plt.legend()
plt.grid(True)
plt.show()
```

## - ùëÇ(ùëõ^2) ùë£ùë† ùëÇ(ùëõ^3)

```python
x = np.arange(1, 10)
y1 = np.power(x, 2)
y2 = np.power(x, 3)

plt.plot(x, y1, '-', label='O(n^2)', color='purple')
plt.plot(x, y2, '-', label='O(n^3)', color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title('O(n^2) vs O(n^3)')
plt.legend()
plt.grid(True)
plt.show()
```

## - ùëÇ(ùëé^ùëõ) ùë£ùë† ùëÇ(ùëõ!)

```python
x = np.arange(1, 11)
a = 4
y1 = np.power(a, x)
y2 = [math.factorial(n) for n in x]

plt.plot(x, y1, '-', label='O(a^n)', color = 'orange')
plt.plot(x, y2, '-', label='O(n!)', color = 'blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('O(a^n) vs O(n!)')
plt.legend()
plt.grid(True)
plt.show()
```

## - ùëÇ(ùëõ!) ùë£ùë† ùëÇ(ùëõ^ùëõ)

```python
x = np.arange(1, 6)
y1 = [math.factorial(n) for n in x]
y2 =  np.power(x, x)

plt.plot(x, y1, '-', label='O(n!)', color = "blue")
plt.plot(x, y2, '-', label='O(n^n)', color = "red")
plt.xlabel('x')
plt.ylabel('y')
plt.title('O(n!) vs O(n^n)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def factorial (n):
    if n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n+1):
            fact *= i
        return fact
```
