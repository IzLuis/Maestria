#### $O(1)$ vs $O(log n)$

#### $O(n)$ vs $O(n log n)$ 

```python
import matplotlib.pyplot as plt
import numpy as np

# Definimos el rango de valores para x
x = np.linspace(1, 100, 100)  # Valores de x desde 0.01 hasta 10, con 100 puntos

# Calculamos los valores de y para cada x
y_lineal = x
y_casi = x * np.log(x)

# Creamos la figura y los ejes
plt.figure()
plt.plot(x, y_lineal, label='O(n)')
plt.plot(x, y_casi, label='O(n log n)')

# Personalizamos el gr谩fico (opcional)
plt.title("Gr谩fica de O(n) vs O(n log n)")
plt.xlabel("Tama帽o de entrada")
plt.ylabel("Tiempo de Ejecuci贸n")
plt.grid(True)
plt.legend()

# Mostramos el gr谩fico
plt.show()
```

#### $O(n^2)$ vs $O(n^3)$

```python
# Definimos el rango de valores para x
x = np.linspace(0, 5, 100)  # Valores de x desde 0.01 hasta 10, con 100 puntos

# Calculamos los valores de y para cada x
y_cuad = x**2
y_cub = x**3

# Creamos la figura y los ejes
plt.figure()
plt.plot(x, y_cuad, label='O(n^2)')
plt.plot(x, y_cub, label='O(n^3)')

# Personalizamos el gr谩fico (opcional)
plt.title("Gr谩fica de O(n^2) vs O(n^3)")
plt.xlabel("Tama帽o de entrada")
plt.ylabel("Tiempo de Ejecuci贸n")
plt.grid(True)
plt.legend()

# Mostramos el gr谩fico
plt.show()
```

#### $O(a^n)$ vs $O(n!)$

```python
from scipy.special import factorial
# Definimos el rango de valores para x
x = np.linspace(0, 10, 100)  # Valores de x desde 0.01 hasta 10, con 100 puntos

# Calculamos los valores de y para cada x
y_exp = 5**x
y_fact = factorial(x)

# Creamos la figura y los ejes
plt.figure()
plt.plot(x, y_exp, label='O(a^n)')
plt.plot(x, y_fact, label='O(n!)')

# Personalizamos el gr谩fico (opcional)
plt.title("Gr谩fica de O(a^n) vs O(n!)")
plt.xlabel("Tama帽o de entrada")
plt.ylabel("Tiempo de Ejecuci贸n")
plt.grid(True)
plt.legend()

# Mostramos el gr谩fico
plt.show()
```

**Haga una tabla donde simule tiempos de ejecuci贸n suponiendo que cada operaci贸n tiene un costo de 1 micro-segundo:**

- **Use como los diferentes tama帽os de entrada $n=100$; $=1000$; $=10000$ y $=100000$.**

- **Note que para algunas f贸rmulas, los n煤meros pueden ser muy grandes.**

```python
import pandas as pd

n1 = 100
n2 = 1000
n3 = 10000
n4 = 100000

df = pd.DataFrame(columns=['Orden','n = 100', 'n = 1000', 'n = 10000', 'n = 100000'], dtype=int)
df
```

Para simular el tiempo de ejecuci贸n de las orden de crecimiento, se declarar谩 una funci贸n donde se ejecute la acci贸n, para poder ejecutar cada una de las n propuestas. Se guardar谩n los resultados en un Dataframe de Pandas.

```python
def O_constante(n):
    t = np.ones_like(n)
    return t

const = pd.Series(["O(1)",
                   O_constante(n1),
                   O_constante(n2),
                   O_constante(n3),
                   O_constante(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```

#### $O(log n)$

```python
def O_log(n):
    t = np.log(n)
    return t

const = pd.Series(["O(log n)",
                   O_log(n1),
                   O_log(n2),
                   O_log(n3),
                   O_log(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```

#### $O(n)$

```python
def O_lineal(n):
    t = n
    return t

const = pd.Series(["O(n)",
                   O_lineal(n1),
                   O_lineal(n2),
                   O_lineal(n3),
                   O_lineal(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```

#### $O(n log n)$

```python
def O_clineal(n):
    t = n * np.log(n)
    return t

const = pd.Series(["O(nlogn)",
                   O_clineal(n1),
                   O_clineal(n2),
                   O_clineal(n3),
                   O_clineal(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```

#### $O(n^2)$

```python
def O_cuad(n):
    t = n ** 2
    return t

const = pd.Series(["O(n^2)",
                   O_cuad(n1),
                   O_cuad(n2),
                   O_cuad(n3),
                   O_cuad(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```

#### $O(n^3)$

```python
def O_cub(n):
    t = n ** 3
    return t

const = pd.Series(["O(n^3)",
                   O_cub(n1),
                   O_cub(n2),
                   O_cub(n3),
                   O_cub(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```

#### $O(a^n)$

En esta funci贸n se utiliza el m贸dulo "decimal" que implementa precisi贸n hasta con 28 d铆gitos.

```python
import decimal

def O_exp(n):
    t = decimal.Decimal(2) ** n
    return t

const = pd.Series(["O(2^n)",
                   O_exp(n1),
                   O_exp(n2),
                   O_exp(n3),
                   O_exp(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```

#### $O(n!)$

En esta funci贸n se usa el m贸dulo "gmpy2" que resuelve operaciones aritm茅ticas con alta precisi贸n, de no usar este m贸dulo, Python marcar铆a los resultados de $1000!$, $10000!$ y $100000!$ como infinitos, lo cual no es un resultado real.

```python
import gmpy2

def O_fact(n):
    t = gmpy2.factorial(n)
    print(t)
    return t

const = pd.Series(["O(n!)",
                   O_fact(n1),
                   O_fact(n2),
                   O_fact(n3),
                   O_fact(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```

#### $O(n^n)$

```python
def O_potexp(n):
    n = decimal.Decimal(n)
    t = n**n
    print(t)
    return t

const = pd.Series(["O(n^n)",
                   O_potexp(n1),
                   O_potexp(n2),
                   O_potexp(n3),
                   O_potexp(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```
