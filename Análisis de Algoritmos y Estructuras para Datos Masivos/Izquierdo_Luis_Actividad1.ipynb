## Introducción

Normalmente, para resolver un problema existen varios algoritmos que podrían ser una solución. Para elegir el mejor camino a tomar, se suele medir la eficiencia de los algoritmos por medio del tiempo de ejecución que estos toman.

> Se puede decir entonces que el algoritmo para un problema requiere un tiempo del orden de $t(n)$ para una función dada $t$, si existe una constante positiva $c$ y una implementación del algoritmo capaz de resolver todos los casos de tamaño $n$ en un tiempo que no sea superior a $ct(n)$ unidades de tiempo. (Bel, 2020, p. 28).

Algunos órdenes se presentan de manera constante en los algoritmos, por lo que ya han sido clasificados en los siguientes:
- De tiempo constante $O(c)$
- De tiempo logarítmico $O(log n)$
- De tiempo lineal $O(n)$
- De tiempo casi lineal $O(n log n)$
- De tiempo polinómico $O(n^a)$
- De tiempo exponencial $O(a^n)$
- De tiempo factorial $O(n!)$
- De orden potencial exponencial $O(n^n)$

En esta actividad haremos comparaciones entre estas órdenes de crecimiento para ver cómo se desempeñan unas contra otras.

#### $O(1)$ vs $O(log n)$

```python
import matplotlib.pyplot as plt
import numpy as np

# Definimos el rango de valores para x
x = np.linspace(1, 100, 100)  # Valores de x desde 0.01 hasta 10, con 100 puntos

# Calculamos los valores de y para cada x
y_const = np.ones_like(x)
y_log = np.log(x)

# Creamos la figura y los ejes
plt.figure()
plt.plot(x, y_const, label='O(1)')
plt.plot(x, y_log, label='O(log n)')

# Personalizamos el gráfico (opcional)
plt.title("Gráfica de O(1) vs O(log n)")
plt.xlabel("Tamaño de entrada")
plt.ylabel("Tiempo de Ejecución")
plt.grid(True)
plt.legend()

# Mostramos el gráfico
plt.show()
```

#### $O(n)$ vs $O(n log n)$ 

```python
import matplotlib.pyplot as plt
import numpy as np

# Definimos el rango de valores para x
x = np.linspace(1, 100, 100)  # Valores de x desde 0.01 hasta 10, con 100 puntos

# Calculamos los valores de y para cada x
y_lineal = x
y_casi = x * np.log(x)

# Creamos la figura y los ejes
plt.figure()
plt.plot(x, y_lineal, label='O(n)')
plt.plot(x, y_casi, label='O(n log n)')

# Personalizamos el gráfico (opcional)
plt.title("Gráfica de O(n) vs O(n log n)")
plt.xlabel("Tamaño de entrada")
plt.ylabel("Tiempo de Ejecución")
plt.grid(True)
plt.legend()

# Mostramos el gráfico
plt.show()
```

#### $O(n^2)$ vs $O(n^3)$

```python
# Definimos el rango de valores para x
x = np.linspace(0, 5, 100)  # Valores de x desde 0.01 hasta 10, con 100 puntos

# Calculamos los valores de y para cada x
y_cuad = x**2
y_cub = x**3

# Creamos la figura y los ejes
plt.figure()
plt.plot(x, y_cuad, label='O(n^2)')
plt.plot(x, y_cub, label='O(n^3)')

# Personalizamos el gráfico (opcional)
plt.title("Gráfica de O(n^2) vs O(n^3)")
plt.xlabel("Tamaño de entrada")
plt.ylabel("Tiempo de Ejecución")
plt.grid(True)
plt.legend()

# Mostramos el gráfico
plt.show()
```

#### $O(a^n)$ vs $O(n!)$

```python
from scipy.special import factorial
# Definimos el rango de valores para x
x = np.linspace(0, 10, 100)  # Valores de x desde 0.01 hasta 10, con 100 puntos

# Calculamos los valores de y para cada x
y_exp = 5**x
y_fact = factorial(x)

# Creamos la figura y los ejes
plt.figure()
plt.plot(x, y_exp, label='O(a^n)')
plt.plot(x, y_fact, label='O(n!)')

# Personalizamos el gráfico (opcional)
plt.title("Gráfica de O(a^n) vs O(n!)")
plt.xlabel("Tamaño de entrada")
plt.ylabel("Tiempo de Ejecución")
plt.grid(True)
plt.legend()

# Mostramos el gráfico
plt.show()
```

#### $O(n!)$ vs $O(n^n)$

```python
from scipy.special import factorial
# Definimos el rango de valores para x
x = np.linspace(0, 3, 100)  # Valores de x desde 0.01 hasta 10, con 100 puntos

# Calculamos los valores de y para cada x
y_exp = x**x
y_fact = factorial(x)

# Creamos la figura y los ejes
plt.figure()
plt.plot(x, y_exp, label='O(n^n)')
plt.plot(x, y_fact, label='O(n!)')

# Personalizamos el gráfico (opcional)
plt.title("Gráfica de O(n^n) vs O(n!)")
plt.xlabel("Tamaño de entrada")
plt.ylabel("Tiempo de Ejecución")
plt.legend()
plt.grid(True)

# Mostramos el gráfico
plt.show()
```

### Inciso 2

**Haga una tabla donde simule tiempos de ejecución suponiendo que cada operación tiene un costo de 1 micro-segundo:**

- **Suponga que cada uno de los ordenes de crecimiento anteriores es una expresión que describe el costo de un algoritmo teniendo en cuenta el tamaño de la entrada del algoritmo n.**

- **Use como los diferentes tamaños de entrada $n=100$; $𝑛=1000$; $𝑛=10000$ y $𝑛=100000$.**

- **Note que para algunas fórmulas, los números pueden ser muy grandes.**

```python
import pandas as pd

n1 = 100
n2 = 1000
n3 = 10000
n4 = 100000

df = pd.DataFrame(columns=['Orden','n = 100', 'n = 1000', 'n = 10000', 'n = 100000'], dtype=int)
df
```

Para simular el tiempo de ejecución de las orden de crecimiento, se declarará una función donde se ejecute la acción, para poder ejecutar cada una de las n propuestas. Se guardarán los resultados en un Dataframe de Pandas.

```python
def O_constante(n):
    t = np.ones_like(n)
    return t

const = pd.Series(["O(1)",
                   O_constante(n1),
                   O_constante(n2),
                   O_constante(n3),
                   O_constante(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```

#### $O(log n)$

```python
def O_log(n):
    t = np.log(n)
    return t

const = pd.Series(["O(log n)",
                   O_log(n1),
                   O_log(n2),
                   O_log(n3),
                   O_log(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```

#### $O(n)$

```python
def O_lineal(n):
    t = n
    return t

const = pd.Series(["O(n)",
                   O_lineal(n1),
                   O_lineal(n2),
                   O_lineal(n3),
                   O_lineal(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```

#### $O(n log n)$

```python
def O_clineal(n):
    t = n * np.log(n)
    return t

const = pd.Series(["O(nlogn)",
                   O_clineal(n1),
                   O_clineal(n2),
                   O_clineal(n3),
                   O_clineal(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```

#### $O(n^2)$

```python
def O_cuad(n):
    t = n ** 2
    return t

const = pd.Series(["O(n^2)",
                   O_cuad(n1),
                   O_cuad(n2),
                   O_cuad(n3),
                   O_cuad(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```

#### $O(n^3)$

```python
def O_cub(n):
    t = n ** 3
    return t

const = pd.Series(["O(n^3)",
                   O_cub(n1),
                   O_cub(n2),
                   O_cub(n3),
                   O_cub(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```

#### $O(a^n)$

En esta función se utiliza el módulo "decimal" que implementa precisión hasta con 28 dígitos.

```python
import decimal

def O_exp(n):
    t = decimal.Decimal(2) ** n
    return t

const = pd.Series(["O(2^n)",
                   O_exp(n1),
                   O_exp(n2),
                   O_exp(n3),
                   O_exp(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```

#### $O(n!)$

En esta función se usa el módulo "gmpy2" que resuelve operaciones aritméticas con alta precisión, de no usar este módulo, Python marcaría los resultados de $1000!$, $10000!$ y $100000!$ como infinitos, lo cual no es un resultado real.

```python
import gmpy2

def O_fact(n):
    t = gmpy2.factorial(n)
    print(t)
    return t

const = pd.Series(["O(n!)",
                   O_fact(n1),
                   O_fact(n2),
                   O_fact(n3),
                   O_fact(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```

#### $O(n^n)$

```python
def O_potexp(n):
    n = decimal.Decimal(n)
    t = n**n
    print(t)
    return t

const = pd.Series(["O(n^n)",
                   O_potexp(n1),
                   O_potexp(n2),
                   O_potexp(n3),
                   O_potexp(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```
