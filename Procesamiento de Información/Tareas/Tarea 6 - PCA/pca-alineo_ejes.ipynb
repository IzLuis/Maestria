# PCA Alineación de ejes

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import decomposition
```

## Creación de los datos

```python
np.random.seed(0)
X = np.random.rand(30, 2) - [0.5, 0.5]
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

Se crean 30 puntos centrados en el origen. Vamos a aplicarles una transformación lineal para modificarlos.

```python
A = [[.4, .5], [.2, 1]]
X = np.array([np.matmul(A, x) for x in X] + np.array([2,2]))
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

Los puntos ya quedaron lejos del origen y con una forma de elipse.

## PCA

```python
pca = decomposition.PCA(n_components=2)
pca.fit(X)
X_t = pca.transform(X)
```

```python
plt.scatter(X_t[:, 0], X_t[:, 1])
plt.show()
```



Al aplicar PCA con el mísmo número de dimensiones, los puntos quedaron centrados en el origen y ya no tienen esa forma de elipse.
