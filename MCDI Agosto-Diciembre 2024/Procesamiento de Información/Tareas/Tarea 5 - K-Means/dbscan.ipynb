# DBSCAN

Se muestran ejemplos del uso de DBSCAN. Las funciones `make_moons`, `make_circles`, `make_classification`, y `make_s_curve` crean conjuntos de puntos con diferentes formas.

```python
import numpy as np
from skimage.io import imread
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons, make_circles, make_classification, make_s_curve
```

```python
color = np.array(["r", "b", "y", "g", "k", "c", "orange", "tomato", "gray", "m"])
```

```python
X, y = make_classification(n_features=2, n_redundant=0, n_informative=2,
                           random_state=1, n_clusters_per_class=1)
plt.scatter(X[:, 0], X[:, 1], c=color[y])
plt.show()
```

Esta es la separación original de los clusters

```python
y_c = DBSCAN(eps=.5, min_samples=2).fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=color[y_c])
plt.show()
```

DBSCAN separa muy bien el conjunto de datos en este caso.

## Medias lunas

```python
X, y = make_moons(n_samples=300, noise=0.05, random_state=0)
plt.scatter(X[:, 0], X[:, 1], c=color[y])
plt.show()
```

```python
y_c = DBSCAN(eps=.2, min_samples=2).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=color[y_c])
plt.show()
```

En el caso de las medias lunas, DBSCAN las separa perfectamente. Note que k-means no lo hubiera podido hacer.

## Muy juntas

```python
X, y = make_moons(n_samples=300, noise=0.1, random_state=1)
plt.scatter(X[:, 0], X[:, 1], c=color[y])
plt.show()
```

```python
y_c = DBSCAN(eps=.15, min_samples=2).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=color[y_c])
plt.show()
```

Se separan bien aún cuando las lunas están muy juntas. Los puntos morados son los que se consideran ruido.

## Otras formas variando el radio

```python
X, y = make_blobs(n_samples=200, centers=10)
plt.scatter(X[:, 0], X[:, 1], c=color[0])
plt.show()
```

## Radio de 5

```python
y_c = DBSCAN(eps=5, min_samples=5).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=color[y_c])
plt.show()
```

Con un radio muy grande, todos los puntos quedan en el mismo cluster.

## Radio = 3

```python
y_c = DBSCAN(eps=3, min_samples=5).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=color[y_c])
plt.show()
```

Con este radio, ya se crearon dos clusters.

## Radio = 2

```python
y_c = DBSCAN(eps=2, min_samples=5).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=color[y_c])
plt.show()
```

Al disminuir el radio, se crean más clusters.

## Radio = 1.5

```python
y_c = DBSCAN(eps=1.5, min_samples=5).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=color[y_c])
plt.show()
```

Los magentas son los puntos marcados como ruido.
