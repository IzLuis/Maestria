# Kmeans

En este notebook se muestra la diferencia al elegir el número de clusters para un conjunto de datos.

La función `make_blobs` crea clusters de manera artifical, en este caso son 5 clusters.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

## Usando la K correcta

```python
X, y = make_blobs(400, 2, centers=5)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
K = 5
km = KMeans(K)
y_km = km.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_km)
plt.show()
```

Al usar la k correcta, se crean los clusters de la manera esperada.

## Con otra K

```python
X, y = make_blobs(400, 2, centers=5)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

## K=4

```python
K = 4
km = KMeans(K)
y_km = km.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_km)
plt.show()
```

Cuando la k es pequeña, k-means pega los clusters.

## K=3

```python
K = 3
km = KMeans(K)
y_km = km.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_km)
plt.show()
```

## K=10

```python
K = 10
km = KMeans(K)
y_km = km.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_km)
plt.show()
```

Cuando la k es grande, k-means separa los clusters.
