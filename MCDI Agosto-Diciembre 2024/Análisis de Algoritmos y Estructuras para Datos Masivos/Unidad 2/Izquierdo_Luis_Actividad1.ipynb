#### $O(1)$ vs $O(log n)$

```python
import matplotlib.pyplot as plt
import numpy as np

# Definimos el rango de valores para x
x = np.linspace(1, 100, 100)

# Calculamos los valores de y para cada x
y_const = np.ones_like(x)
y_log = np.log(x)

# Creamos la figura y los ejes
plt.figure()
plt.plot(x, y_const, label='O(1)')
plt.plot(x, y_log, label='O(log n)')

# Personalizamos el gr谩fico (opcional)
plt.title("Gr谩fica de O(1) vs O(log n)")
plt.xlabel("Tama帽o de entrada")
plt.ylabel("Tiempo de Ejecuci贸n")
plt.grid(True)
plt.legend()

# Mostramos el gr谩fico
plt.show()
```

En la figura anterior podemos observar como el tiempo de ejecuci贸n para ambos casos difiere mucho.
En el caso del tiempo constante vemos como no se afecta por el tama帽o de entrada, a diferencia del tiempo logar铆tmico que aumenta el tiempo de ejecuci贸n con el tama帽o de entrada

#### $O(n)$ vs $O(n log n)$ 

```python
x = np.linspace(1, 100, 100) 

y_lineal = x
y_casi = x * np.log(x)

plt.figure()
plt.plot(x, y_lineal, label='O(n)')
plt.plot(x, y_casi, label='O(n log n)')
plt.title("Gr谩fica de O(n) vs O(n log n)")
plt.xlabel("Tama帽o de entrada")
plt.ylabel("Tiempo de Ejecuci贸n")
plt.grid(True)
plt.legend()

plt.show()
```

En esta comparaci贸n vemos como, a pesar de lucir diferente, si analizamos bien podemos observar que las l铆neas son muy parecidas, ambos tiempos de ejecuci贸n son directamente proporcionales al tama帽o de entrada, esto es esperado ya que la pista la tenemos en los nombres: tiempo lineal y tiempo casi lineal.

#### $O(n^2)$ vs $O(n^3)$

```python
x = np.linspace(0, 5, 100)  

y_cuad = x**2
y_cub = x**3

plt.figure()
plt.plot(x, y_cuad, label='O(n^2)')
plt.plot(x, y_cub, label='O(n^3)')
plt.title("Gr谩fica de O(n^2) vs O(n^3)")
plt.xlabel("Tama帽o de entrada")
plt.ylabel("Tiempo de Ejecuci贸n")
plt.grid(True)
plt.legend()
plt.show()
```

En esta gr谩fica podemos observar como la operaci贸n de tiempo cuadr谩tica parece ser m谩s eficiente que la operaci贸n de tiempo c煤bica, ya que al aumentar el tama帽o de entrada, el tiempo de ejecuci贸n no se dispara al mismo nivel que la c煤bica, la cual, crece bastante con el mismo tama帽o de entrada.

#### $O(a^n)$ vs $O(n!)$

```python
#Importamos la funcion factorial de scipy
from scipy.special import factorial

x = np.linspace(0, 10, 100)

y_exp = 5**x
y_fact = factorial(x)

plt.figure()
plt.plot(x, y_exp, label='O(a^n)')
plt.plot(x, y_fact, label='O(n!)')
plt.title("Gr谩fica de O(a^n) vs O(n!)")
plt.xlabel("Tama帽o de entrada")
plt.ylabel("Tiempo de Ejecuci贸n")
plt.grid(True)
plt.legend()

plt.show()
```

En esta gr谩fica podemos observar como las funciones parecen ser bastante parecidas, ya que podemos notar como la operaci贸n de tiempo factorial comienza a subir de la misma manera que la de tiempo exponencial.

#### $O(n!)$ vs $O(n^n)$

```python
x = np.linspace(0, 3, 100)

y_exp = x**x
y_fact = factorial(x)

plt.figure()
plt.plot(x, y_exp, label='O(n^n)')
plt.plot(x, y_fact, label='O(n!)')
plt.title("Gr谩fica de O(n^n) vs O(n!)")
plt.xlabel("Tama帽o de entrada")
plt.ylabel("Tiempo de Ejecuci贸n")
plt.legend()
plt.grid(True)

plt.show()
```

Gracias a esta gr谩fica podemos observar que la operaci贸n de tiempo factorial, a pesar de ser una de las m谩s tardadas, puede ser much铆simo mejor que aquella de tiempo potencial exponencial, ya que esta 煤ltima se dispara en tiempo de ejecuci贸n con el mismo tama帽o de entrada.

```python
#Importamos pandas para crear la tabla
import pandas as pd

#Creamos las variables de n que ser谩n utilizadas a lo largo del c贸digo
n1 = 100
n2 = 1000
n3 = 10000
n4 = 100000

#Creamos el dataframe con las columnas para guardar los datos
df = pd.DataFrame(columns=['Orden','n = 100', 'n = 1000', 'n = 10000', 'n = 100000'], dtype=int)
df
```

Para simular el tiempo de ejecuci贸n de las orden de crecimiento, se declarar谩 una funci贸n donde se ejecute la acci贸n, para poder ejecutar cada una de las n propuestas. Se guardar谩n los resultados en un Dataframe de Pandas.

```python
def O_constante(n):
    t = np.ones_like(n)
    return t

const = pd.Series(["O(1)",
                   O_constante(n1),
                   O_constante(n2),
                   O_constante(n3),
                   O_constante(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```

#### $O(log n)$

```python
def O_log(n):
    t = np.log(n)
    return t

const = pd.Series(["O(log n)",
                   O_log(n1),
                   O_log(n2),
                   O_log(n3),
                   O_log(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```

```python
def O_lineal(n):
    t = n
    return t

const = pd.Series(["O(n)",
                   O_lineal(n1),
                   O_lineal(n2),
                   O_lineal(n3),
                   O_lineal(n4)
                   ], index=df.columns)
df.loc[len(df)] = const
df
```
